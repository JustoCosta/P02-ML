```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PathTrain = 'E-Commerce_train.xlsx'
PathTest = 'E-Commerce_test.xlsx'
```

VALORES DE CADA CAMPO 

ID: identificador del registro de orden (valor entero).
Warehouse_block: Almacén de distribución de donde salió la orden (A a F).
Mode_of_Shipment: Medio de transporte (Flight, Road, Ship).
Customer_care_calls: Número de llamadas a atención al cliente que hubo por esa orden. (valores enteros del 2 al 7)
Customer_rating: Puntaje del cliente (valores enteros 1 al 5).
Cost_of_the_Product: Costo del producto (valor numérico entero de 96 a 310).
Prior_purchases: Número de compras previas realizadas por el cliente (valor numérico entero de 2 a 10).
Product_importance: Nivel de importancia del producto (low, medium, high).
Gender: Género del comprador (F, M).
Discount_offered: Porcentaje de descuento ofrecido por esa compra (valor numérico entero de 1 a 65):
Weight_in_gms: Peso del paquete de la orden, en gramos (valor numérico entero de 1001 a 7846).
Reached.on.Time_Y.N: Información sobre la llegada del paquete a destino (1 si llegó a tiempo, 0 si no llegó a tiempo).

```python
#Cargamos el archivo de entrenamiento formato excel a Dataframe
DataTrain = pd.read_excel(PathTrain, sheet_name = 'Sheet1')
DataTrain
```

```python
#Cambiamos los tipos de variables de las columnas con campos que contienen string a numericos.
DataTrain.Mode_of_Shipment = DataTrain.Mode_of_Shipment.replace({'Road': 1 , 'Flight': 2 , 'Ship' : 3})
DataTrain.Product_importance = DataTrain.Product_importance.replace({'low': 1 , 'medium': 2 , 'high' : 3})
DataTrain.Warehouse_block = DataTrain.Warehouse_block.replace({'A': 1 , 'B': 2 , 'C' : 3 , 'D': 4 , 'E': 5 , 'F' : 6})
DataTrain
```

```python
#No tenemos en cuenta las columnas 'ID' y 'Genger' de los atributos ya que no aportan influencia en el resultado de manera directa. 
X = DataTrain[['Warehouse_block','Mode_of_Shipment', 'Customer_care_calls', 'Customer_rating','Cost_of_the_Product','Prior_purchases','Product_importance','Discount_offered' , 'Weight_in_gms']]
y = DataTrain[['Reached.on.Time_Y.N']]
```

```python
#Importamos el separador
from sklearn.model_selection import train_test_split

#Separamos el %30 de los datos para el test y lo restante lo entrenamos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Instanciamos un objeto de la clase DecisionTreeClassifier

Arbol = DecisionTreeClassifier(max_depth = 4, random_state = 0) 
```

```python
# Entrenamos el modelo

Arbol.fit(X_train.values,y_train.values)
```

```python
# Predecimos los atributos apartados para el testeo

y_pred = Arbol.predict(X_test.values)
```

```python
from sklearn.metrics import accuracy_score , recall_score

Accuracy = accuracy_score(y_pred,y_test)
Recall = recall_score(y_pred,y_test)

print(Accuracy , Recall)
```
